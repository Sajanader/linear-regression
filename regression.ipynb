# Linear Regression 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

- Load the data you receive into a Pandas DataFrame.
- Show the first five rows of the data set.

```python
housing = pd.read_csv('./USA_Housing.csv')
housing.head(5)
```

- Show the description and the info of the data set.

```python
housing.describe()
```

```python
housing.info()
```

```python
x = housing['Avg. Area Income'].values.reshape(-1,1)

y = housing['Price'].values.reshape(-1,1)
y, x
```

- Using a regression model, split your data into train and test data.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=100)
```

```python

%matplotlib inline
plt.scatter(x_train, y_train, color='green')
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.title('The relationship between Avg. Area Income and price')
plt.show()
```

- Fit your training split to the regression model.
- Show your regression models score.

```python
model = LinearRegression()
model.fit(x_train, y_train)
model
y_predict = model.predict(x_test)

```

```python
print(f'Test accuracy{model.score(x_test, y_test)*100}%')
print(f'Train accuracy {model.score(x_train, y_train) * 100} %')
```

- Draw at least three conclusions from your regression model.

```python

%matplotlib inline
plt.scatter(x_train, y_train, color='purple')
plt.scatter(x_test, y_test, color='y')
plt.plot(x_test, y_predict, color='pink')
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.title('The relationship between Avg. Area Income and price')
plt.show()
```
